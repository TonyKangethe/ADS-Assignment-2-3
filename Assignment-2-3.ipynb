# IBM ATTRITION

Attrition is the state at which employees exit a company either through resignation, retirement or death. This analysis is an overview of attrition with respect to job role, monthly income as well as distance from home

## Start by importing the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import the dataset to be used.

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data
```

An overview of what the dataset encompases i.e Number of columns, Datatype of the columns, Number of entries

```python
data.info()
```

Get a statistical overview of all numerical columns within the dataset

```python
data.describe()
```

```python
data.isna().sum() #check for null values with the dataframe
```

Since the dataset does not indicate the rate of exit (attrition), it can be calculates as show below

```python
num_attrition  = len(data[data.Attrition=='Yes']) #Number of entries within the dataset with the Attrition column set to Yes

num_retained = len(data[data.Attrition=='No'])  #Number of entries within the dataset with the Attrition column set to No

total_staffs = len(data)  #Total number of entries in the dataset

df = [num_attrition, num_retained]
labels = ['YES', 'NO']
colors = sns.color_palette('pastel')[0:2]
plt.figure(figsize=(10,8))
plt.pie(df,
       labels=labels,
       colors=colors,
       autopct='%.0f%%')
print('The Rate of Attrition at the Company is ', round(num_attrition/total_staffs*100),'%')
plt.show()

```

# Question One
Show a breakdown of distance from home by job role and attrition.

By grouping our dataset by Job role and attrition with respect to cumulative distance from home. The Sales Executive role had the highest distance from home followed by Research Scientist and Laboratory Technician roles respectively. As show by the two figures below all the three roles had the highest total distance from home for employees who had left the company

```python
role_attrition = data.groupby(by=['JobRole', 'Attrition']).sum()['DistanceFromHome'].reset_index(name='TotalDistance')

role_attrition
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data = role_attrition,
           y = 'JobRole',
           x = 'TotalDistance',
           hue = 'Attrition')
plt.title('The Total Distance from home for Various Job Roles')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(data = data,
             y = data.JobRole,
             hue = 'Attrition')
plt.title('A distribution of the Distance from home for Various Job Roles')
plt.show()
```

# Question Two

Compare average monthly income by education and attrition

From the below figures employees with an education background in Human Resources had highest average monthly income as well as the lowest average monthly income for employees who has exited the company. However, the Marketing education field had the second highest average monthly income, but had the highest average monthly income for employees who left the company.

```python
education_vs_income = data.groupby(by=['EducationField','Attrition']).mean()['MonthlyIncome'].reset_index(name='AverageIncome')
education_vs_income
```

```python
plt.figure(figsize=(15,12))
sns.barplot(data = education_vs_income,
           y = 'EducationField',
           x = 'AverageIncome',
           hue = 'Attrition')
plt.title('The Average Monthly Income for the Various Education Fields')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.boxplot(data = data,
            y = 'EducationField',
            x = 'MonthlyIncome',
            hue = 'Attrition')
plt.title('A Distribution of Monthly Income for the Various Education Fields')
plt.show()
```
